```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
#Load model parameters
model = torch.load('model.pt', weights_only=False)
model.eval()
```

```python
# Read training data
t_train = []
train_losses = []
with open('train_data/train_losses.dat', 'r') as f:
    for line in f:
        v = list(map(float, line.split()))
        t_train.append(v[0])
        train_losses.append(v[1])
t_test = []
test_losses = []
with open('train_data/test_losses.dat', 'r') as f:
    for line in f:
        v = list(map(float, line.split()))
        t_test.append(v[0])
        test_losses.append(v[1])
```

```python
plt.plot(t_train,train_losses,zorder=5)
plt.scatter(t_test,test_losses,color='r',zorder=15)
plt.style.use('fivethirtyeight')
plt.xlabel('Iterations')
plt.ylabel('Loss')
```

```python
t_train = []
sindy_weights = [[] for i in range(10)]
with open('train_data/sindy_weights.dat', 'r') as f:
    for line in f:
        v = list(map(float, line.split()))
        t_train.append(v[0])
        for i in range(1,len(v)):
            sindy_weights[i-1].append(v[i])
```

```python
for i in range(10):
    plt.plot(t_train,sindy_weights[i])
plt.xlabel('Iterations')
plt.ylabel('sindy_weight')
```
